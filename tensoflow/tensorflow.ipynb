```python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train[0]
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train[0]
```

```python
x_train[:1]
```

untrainded model

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10) # output class 10
])
```

```python
model.summary()
```

```python
predictions = model(x_train[:1])
predictions
```

```python
predictions = model(x_train[0:1]).numpy() # slicing 0:1
predictions
```

```python
predictions1 = model(x_train[:1]).numpy() # 
predictions
```

```python
predictions = model(x_train[:2]).numpy() # 0:2 slice for
predictions
```

```python
tf.nn.softmax(predictions1).numpy()
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
loss_fn(y_train[:1], predictions1).numpy()
```

```python
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

Train and evaluate your model

```python
model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test,  y_test, verbose=2)
```

```python
probability_model = tf.keras.Sequential([
  model,
  tf.keras.layers.Softmax()
])
```

```python
probability_model.summary()
```
